```python
import os, sys

project_root = os.path.abspath(os.path.join(os.getcwd(), ".."))
if project_root not in sys.path:
    sys.path.append(project_root)

import matplotlib.pyplot as plt
import pandas as pd

from src.data_cleaning import prepare_core_frame
from src.forecasting import prepare_country_series, arima_forecast, PROPHET_AVAILABLE, prophet_forecast
from src.utils import print_section
```

```python
df = prepare_core_frame()
countries = sorted(df["country"].unique())
len(countries), countries[:10]
```

```python
country = "Germany"   # change to any country name from the list
print_section(f"Forecasting for {country}")

series = prepare_country_series(df, country)
series.tail()
```

```python
plt.figure(figsize=(10, 4))
plt.plot(series.index, series.values, marker="o")
plt.title(f"{country} CO₂ emissions over time")
plt.xlabel("Year")
plt.ylabel("Emissions")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
steps = 15
mean, lower, upper = arima_forecast(series, steps=steps)

forecast_years = range(int(series.index.max()) + 1, int(series.index.max()) + 1 + steps)
forecast_df = pd.DataFrame(
    {
        "year": list(forecast_years),
        "forecast": mean.values,
        "lower": lower.values,
        "upper": upper.values,
    }
)

forecast_df
```

```python
plt.figure(figsize=(10, 5))
plt.plot(series.index, series.values, label="History", marker="o")
plt.plot(forecast_df["year"], forecast_df["forecast"], label="Forecast", marker="o")
plt.fill_between(forecast_df["year"], forecast_df["lower"], forecast_df["upper"], alpha=0.2, label="Confidence interval")
plt.title(f"{country} – ARIMA forecast")
plt.xlabel("Year")
plt.ylabel("Emissions")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
