```python
import os
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Optional but very useful: auto-reload changed modules
%load_ext autoreload
%autoreload 2

# Make plots readable
plt.rcParams["figure.figsize"] = (10, 4)
plt.rcParams["axes.grid"] = True
plt.rcParams["axes.labelsize"] = 12
plt.rcParams["axes.titlesize"] = 14

# Add project root to path
project_root = os.path.abspath(os.path.join(os.getcwd(), ".."))
if project_root not in sys.path:
    sys.path.append(project_root)

from src.data_cleaning import prepare_core_frame
from src.forecasting import prepare_country_series
from src.model_comparison import evaluate_models_for_country
```

```python
df = prepare_core_frame()
print("Data shape:", df.shape)
df.head()
```

```python
countries = sorted(df["country"].unique())
print("Number of countries:", len(countries))
countries[:20]
```

```python
country = "Germany"  # change as needed

series = prepare_country_series(df, country)
print(country, "series length:", len(series))
series.head(), series.tail()
```

```python
test_years = 8   # how many last years as test
val_years = 5    # how many just before that as validation

results = evaluate_models_for_country(
    series,
    country=country,
    test_years=test_years,
    val_years=val_years,
)

results
```

```python
arima_res = results["ARIMA"]

metrics_df = pd.DataFrame(
    {
        "MAE (tons)": [arima_res["MAE"]],
        "RMSE (tons)": [arima_res["RMSE"]],
        "MAPE (%)": [arima_res["MAPE"]],
        "SMAPE (%)": [arima_res["SMAPE"]],
        "order (p,d,q)": [str(arima_res["order"])],
    },
    index=["ARIMA"],
)

metrics_df.round(3)
```

```python
test = series.iloc[-test_years:]
years_test = test.index.values
pred = np.asarray(arima_res["pred"], dtype=float)

plt.figure()
plt.plot(test.index, test.values, marker="o", label="Actual")
plt.plot(years_test, pred, marker="o", label="ARIMA")

plt.title(f"Actual vs ARIMA forecast – {country}")
plt.xlabel("Year")
plt.ylabel("CO₂ emissions (tons)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
errors = np.abs(test.values - pred)

plt.figure()
plt.plot(years_test, errors, marker="o", label="ARIMA error")
plt.title(f"Absolute error per year – {country}")
plt.xlabel("Year")
plt.ylabel("Absolute error (tons)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
test_years = 8
val_years = 5

rows = []

for c in countries:
    s = prepare_country_series(df, c)
    if len(s) <= test_years + val_years + 5:
        # Too short for proper split, skip
        continue

    try:
        res = evaluate_models_for_country(s, c, test_years=test_years, val_years=val_years)
        arima_res = res["ARIMA"]
        rows.append(
            {
                "country": c,
                "MAE": arima_res["MAE"],
                "RMSE": arima_res["RMSE"],
                "MAPE": arima_res["MAPE"],
                "SMAPE": arima_res["SMAPE"],
                "order": str(arima_res["order"]),
            }
        )
    except Exception as e:
        print(f"Skipping {c}: {e}")

summary_df = pd.DataFrame(rows)
summary_df.head()
```

```python
summary_sorted = summary_df.sort_values("RMSE").reset_index(drop=True)
summary_sorted.head(15)
```

```python
summary_sorted.tail(15)
```
