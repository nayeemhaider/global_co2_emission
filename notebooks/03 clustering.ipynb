```python
import os, sys

project_root = os.path.abspath(os.path.join(os.getcwd(), ".."))
if project_root not in sys.path:
    sys.path.append(project_root)

import matplotlib.pyplot as plt
import seaborn as sns

from src.clustering import run_clustering, build_country_features
from src.data_cleaning import prepare_core_frame
from src.utils import print_section
```

```python
df_clustered["cluster"].value_counts()
```

```python
sns.pairplot(
    df_clustered,
    vars=["total_emission", "mean_emission", "emission_trend"],
    hue="cluster",
    corner=True
)
plt.show()
```

```python
cluster_profiles = (
    df_clustered
    .groupby("cluster")[["total_emission", "mean_emission", "emission_trend"]]
    .mean()
    .round(2)
)
cluster_profiles
```
